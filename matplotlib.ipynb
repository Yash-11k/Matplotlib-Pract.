```python
import matplotlib.pyplot as plt 
#pyplot is a collection of function that make matplotlib worka as MATLAB 
```

```python
# 1 Create first char 
```

```python
x= [ 1,2,3,4,5]
y= [10,20,35,35,45]

plt.plot(x,y) #line char 
plt.show() # dispaly the chart to use this it is a good practise 
```

```python
#Customize Chart 
```

```python
plt.figure(figsize=(8,4)) #to change the size of a chart (most present in inches wrt x,y )
plt.plot(x,y, color = 'red',  marker= 'o', linestyle = 'dashed', linewidth=2, markersize= 12)



plt.title("Title of the grapht")
plt.xlabel("X-axis ka label")
plt.ylabel("Y-axis ka label")



plt.show()
```

```python
#ek hi chart m 2 lines dikhate hai 
#Advanced Line Chart and Legends 
```

```python
x = [9,8,7,6,5,4]
y1= [1,2,3,4,5,6]
y2= [1,4,9,16,25,36]


plt.plot(x,y1, label= 'Sales 2023')
plt.plot(x,y2, label= 'Sales 2024')

plt.title("YoY Sales")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.legend()
plt.show()

```

```python
x= [1,2,3,4,5]
y = [10,20,30,40,50]

plt.bar(x,y)
plt.title("THe Bar Chart")
plt.xlabel("X ka label")
plt.ylabel("Y ka label")





plt.show()
```

```python
#Histogram
```

```python
#used for distribution analysis 
#data 
import random
data = [random.randint(1,100) for _ in range(1000)]
plt.hist(data , bins = 20) # ye bins granulity dikhat ahai , kitne parts m toda hai isko
plt.title("Histogram Example") #continous data hota hai isme
plt.show()
```

```python
# Pie Chart(whole data ka part ya relationship dikhana ho)
```

```python
x= ['A', 'B', 'C', 'D', 'E']
sales=  [10,20,30,40,50]

plt.pie(sales, labels= x, autopct= '%1.1f%%', startangle= 90)
plt.title("Pie chart Exaample")
plt.show()
```

```python
#Scatter Plot- used to find relationship between variables just similar as clustering
```

```python
y1 = [10,20,30,40,50]
y2 = [20,30,40,50,60]

plt.scatter(y1,y2)
plt.title("Scatter Plot Example")
plt.show()
```

```python
# subplots daily sales , weekly sales , slaes by category , size bta rhae ho kitne charts rows aur coluums 
```

```python
# data-1 - Bar Chart
categories = ['A', 'B', 'C', 'D', 'E']
sales=  [10,20,30,40,50]

# data-2 - Scatter Plot
y1= [ 10,20,30,40,50]
y2= [ 20,30,40,50,60]


plt.figure(figsize=(10,4))


#first plot - bar chart 

plt.subplot(2,1,2) #row, columm , position
plt.bar(categories, sales)
plt.title("Weekly Sales")
plt.xlabel("Week Days")
plt.ylabel("Sales")


#second plot - scatter chart 

plt.subplot(2,1,1) # row, columm , position
plt.scatter(y1, y2)
plt.title("User Examples")
plt.xlabel("User 1")
plt.ylabel("User 2 ")


plt.show()
```

```python
#Matplotlib with pandas - real data 
```

```python
# create df 
import pandas as pd 

data = {
    'Month' : ['Jan', 'Feb', 'Mar', 'Apr'],
    'Sales' : [ 12000, 11000, 13000, 25000]
}

df = pd.DataFrame(data) #capital likhan hai dataframe
df
```

```python
plt.bar(df['Month'], df['Sales'])
plt.title("Matplotib with Pandas")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
# Save Charts/plots

```

```python
plt.bar(df['Month'], df['Sales'])
plt.title("Matplotib with Pandas")
plt.xlabel("Month")
plt.ylabel("Sales")


plt.savefig("Monhtly_sales_from_df.png")
plt.show()
```
